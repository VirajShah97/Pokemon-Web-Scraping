### Step 1.2 and 1.3 
### Scrape the photo URLs for each pokemon in sample2. Display the photos for all of the sample2 pokemon in your Jupyter notebook

```python
def imageurl():
    for i in df1['Name']:
            url = 'https://pokemondb.net/pokedex/' + i
            content = requests.get(url).content
            soup1 = BeautifulSoup(content,'html.parser')
            image_tags = soup1.find_all('img', {'src':re.compile('.jpg')})
            for image_tag in image_tags:
                j = image_tag.get('src')
                print("The image URL for {} is : {}" .format(i,j))
                display(Image(url = j, unconfined=True, height = 100, width = 100))
```

## Exercise #2

### Step 2.1
### Using BeautifulSoup, scrape the Location table for Bulbasaur properly and add it to a DataFrame

```python
tables = pd.read_html(requests.get("https://pokemondb.net/pokedex/bulbasaur",headers={'User-agent': 'Mozilla/5.0'}).text)[-2]
x = pd.DataFrame(tables)
x = x.rename({0:'Video_game',1:'Location'}, axis = 1)
x['Location'] = np.where(x['Location'] == 'Location data not yet available', np.nan, x['Location'])
x = x.T
```

### When multiple columns are combined, separate the columns and duplicate the location information

```python
new_list = []
for i in x.iloc[0]:
    new_list.append(re.sub(r"([A-Z])", r" \1", i).split())

table_dict = {'Video_game': [],'Location': []}

for j in range(len(new_list)):
    for k in new_list[j]:
        table_dict['Video_game'].append(k)
        table_dict['Location'].append(x.iloc[1][j])
df2 = pd.DataFrame(table_dict)
df2 = df2.T
df2.columns = df2.iloc[0]
df2 = df2.iloc[1:]
```

### Step 2.2

###  Generalize Step 2.1 for all pokemon in the original sample (the sample from Step 3.1 from HW #4). Append all the data together. If information is not available for a pokemon in a particular game, the column should contain NaN.

```python
df3 = df.loc[df['sample'] == 1]
df3 = df3[df3['Name'] != 'Meltan']
```

```python
my_list = []
for x in df3['URL']:
    time.sleep(0.01)
    url1 = 'https://pokemondb.net' + x
    poke_scrape = requests.get(url1)
    soup1 = BeautifulSoup(poke_scrape.text,'html.parser')
    pokemon_table = soup1.find_all(name = 'table', class_='vitals-table')
    pokemon_gen = pd.read_html(str(pokemon_table))[-2]
    pokemon_list = pokemon_gen.values.tolist()
    table_loc = pd.DataFrame(pokemon_list)
    table_loc = table_loc.T
    video = []
    dict_table = {'Video_game':[],'Location':[]}
    for a in table_loc.iloc[0]:
        video.append(re.sub( r"([A-Z])", r" \1", a).split())
    for y in range(len(video)):
        for z in video[y]:
            dict_table['Video_game'].append(z)
            dict_table['Location'].append(table_loc.iloc[1][y])
    my_list.append(pd.DataFrame(dict_table))
    
new_df = pd.concat(my_list,axis = 1)
new_df = new_df.T
display_columns = new_df.iloc[0]
new_df = new_df[1:]
new_df = new_df.iloc[::2]
new_df = np.where(new_df == 'Location data not yet available', np.nan, new_df)
display_table = pd.DataFrame(new_df)
display_table.columns = display_columns
```
